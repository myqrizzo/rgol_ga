```python
import numpy as np
import pandas as pd
from viz import show_field
from MPGeneticSolver import MPGeneticSolver
```

```python
test_df = pd.read_csv("resources/test.csv", index_col='id')
rf_pred = pd.read_csv("s_0.4473684210526315.csv", index_col='id')
```

```python
for i, (id, (idx, row)) in enumerate(zip(test_df.index, test_df.iterrows())):
    delta, Y = row.values[0], row.values[1:].reshape((20, 20)).astype('uint8')
    if delta == 1:
        solver = MPGeneticSolver(initialization_strategy=rf_pred.iloc[i].values, patience=50, n_proc=10)
        results = solver.solve(Y, delta, return_all=True)
        boards = [board for (board, score) in results]
        avg_board = np.zeros_like(boards[0])
        for b in boards:
            avg_board += b
        avg_board = avg_board / 10
        avg_board[avg_board >= 0.5] = 1
        avg_board[avg_board < 0.5] = 0
        rf_pred.iloc[i] = avg_board.astype(np.uint8).ravel()
    if i % 500 == 0:
        print("i =", i)
```

```python
rf_pred.to_csv("d1_enhanced_avg.csv")
```
